# 2013

```python
y2013_df = pd.read_csv('./raw_data/201306-citibike-tripdata.csv', encoding='utf8')
y2013_df.head()
```

```python
import math
y2013_df['tripduration_min'] = (y2013_df['tripduration']/60).round(2)
y2013_df['birth year'] =  y2013_df['birth year'].map(lambda x: 'N/A' if math.isnan(x) else round(x))
y2013_df.head()
```

```python
y2013_df['gender'].replace(0, 'Other',inplace=True)
y2013_df['gender'].replace(1, 'Male',inplace=True)
y2013_df['gender'].replace(2, 'Female',inplace=True)
y2013_df['year']='2013'
y2013_df['rideable_type'] = 'NaN'
y2013_df.head()
```

```python
y2013_df.to_csv('2013.csv')  
```

```python
merged_df = y2013_df
```

```python
y2021_df = pd.read_csv('./raw_data/202105-citibike-tripdata.csv', encoding='utf8')
y2021_df.head()
```

```python
import time
import datetime
s = "2021-05-23 16:51:00"
time.mktime(datetime.datetime.strptime(s, "%Y-%m-%d %H:%M:%S").timetuple())
```

```python
s = "2021-05-23 18:38:21"
time.mktime(datetime.datetime.strptime(s, "%Y-%m-%d %H:%M:%S").timetuple())
```

```python
def gettimestamp (s):
    return time.mktime(datetime.datetime.strptime(s, "%Y-%m-%d %H:%M:%S").timetuple())
```

```python
y2021_df['duration_min'] = y2021_df.apply(
    lambda row: round(
                    (gettimestamp(row['ended_at'])- gettimestamp(row['started_at']))/60
                ),
    axis =1
)
```

```python
y2021_df['year']='2021'
y2021_df.head()
```

```python
y2021_df.rename(columns = {'ride_id':'bikeid', 'started_at':'starttime', 'ended_at':'stoptime', 
                           'start_station_name':'start station name', 'start_stattion_id':'start station id','start_station_id':'start station id', 'end_station_name': 'end station name', 'end_station_id': 'end station id',
                           'start_lat': 'start station latitude', 'start_lng':'start station longitude',
                           'end_lat': 'end station latitude', 'end_lng':'end station longitude', 
                           'member_casual':'usertype', 'duration_min': 'tripduration_min'
                          }, inplace = True)
y2021_df.head()
```

```python
y2021_df['gender'] = 'Other'
y2021_df['birth year'] = 'NaN'
y2021_df['tripduration'] = 'NaN'
```

```python
y2021_df = y2021_df[['tripduration', 'starttime', 'stoptime', 'start station id',
       'start station name', 'start station latitude',
       'start station longitude', 'end station id', 'end station name',
       'end station latitude', 'end station longitude', 'bikeid', 'usertype',
       'birth year', 'gender', 'tripduration_min', 'year', 'rideable_type']]
y2021_df.head()
```

```python
y2021_df.to_csv('2021.csv')  
```

```python
print(merged_df.shape)
merged_df = merged_df.append(y2021_df, ignore_index=True)
print(merged_df.shape)
```

```python
print(merged2_df.shape)
merged2_df = merged2_df.append(y2021_df, ignore_index=True)
print(merged2_df.shape)
```
